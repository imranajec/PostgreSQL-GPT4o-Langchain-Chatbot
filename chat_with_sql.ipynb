```python
import os 
from langchain.agents import *
from langchain.llms import OpenAI
from langchain.sql_database import SQLDatabase
from langchain.agents.agent_toolkits import SQLDatabaseToolkit
from langchain.agents import AgentExecutor
```

```python
db_user = "postgres"
db_password = ""
db_host = "localhost"
db_name ="Olympics"
```

```python
db_uri = f"postgresql://{db_user}:{db_password}@{db_host}/{db_name}"
```

```python
db = SQLDatabase.from_uri(f"postgresql://{db_user}:{db_password}@{db_host}/{db_name}")
```

```python
from langchain.chat_models import ChatOpenAI
llm = ChatOpenAI(model_name = "gpt-3.5-turbo")
```

```python
toolkit = SQLDatabaseToolkit(db=db,llm=llm)
```

```python
toolkit = SQLDatabaseToolkit(db=db, llm=llm)
agent_executor = create_sql_agent(
    llm=llm,
    toolkit=toolkit,
    verbose=True,
    handle_parsing_errors=True  # Add this to handle parsing errors
)
```

```python
try:
    response = agent_executor.run("How many tables do we have?")
    print("Response:", response)
except Exception as e:
    print("An error occurred:", str(e))
```

```python
try:
    response = agent_executor.run("find the last 5 Olympic Games held as per the dataset")
    print("Response:", response)
except Exception as e:
    print("An error occurred:", str(e))
```
